## Librerias

```python
from pyspark.sql import functions as f
```

## Variables

```python
%run ../00-Modulos/n_modulo_variables.ipynb
```

## Spark Session

```python
spark = spark_session_hive("Lakehouse-Iceberg")
```

## Extracción

```python
df_sales_transactions = spark.read.table(f'{vSchemaLand}.sales_transactions')
```

## Transformaciones

```python
df_sales_transactions = df_sales_transactions.withColumn("TRANSACTION_DATE_ID", f.date_format(f.col("TRANSACTION_DATE"), "yyyyMMdd").cast("int"))
df_final = df_sales_transactions.select("TRANSACTION_DATE_ID", *[c for c in df_sales_transactions.columns if c != "TRANSACTION_DATE_ID"])
```

## Carga

```python
df_final.writeTo(f'{vSchemaBron}.sales_transaction').overwrite(f.lit(True))
```

## Cierre de SparkSession

```python
stop_session(spark)
```
