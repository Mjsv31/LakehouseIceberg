### Schema

```python
# - Crea el schema si no existe
spark.sql(f"""
CREATE SCHEMA IF NOT EXISTS {vSchemaGold}
LOCATION "{vWarehouse}/gold"
""")
```

### Dimension de Clientes

```python
# - Crea tabla
spark.sql(f"""
CREATE TABLE IF NOT EXISTS {vSchemaGold}.DIM_CUSTOMERS
(
    CUST_SK          BIGINT     NOT NULL COMMENT 'Llave Subrogada del cliente',
	CUST_ID          INT        NOT NULL COMMENT 'Identificador del cliente',
	CUST_NAME        STRING              COMMENT 'Nombre del cliente',
	CUST_ADDRESS     STRING              COMMENT 'Direccion del cliente',
	CUST_CITY        STRING              COMMENT 'Ciudad del cliente',
	CUST_STATE       STRING              COMMENT 'Estado del cliente',
	CUST_ZIP_CODE    INT                 COMMENT 'Codigo postal del cliente',
	CUST_COUNTRY     STRING              COMMENT 'Pais del cliente',
	CUST_PHONE_NMBR  STRING              COMMENT 'Numero telefonico del cliente',
	CUST_GENDER      STRING              COMMENT 'Genero del cliente',
	CUST_AGE_GROUP   STRING              COMMENT 'Grupo de edad al que pertenece el cliente',
	CUST_INCOME      DECIMAL(12,2)       COMMENT 'Ingreso del cliente',
	CUST_E_MAIL      STRING              COMMENT 'Correo electronico del cliente',
	CUST_AGE         INT                 COMMENT 'Edad del cliente'
)
USING ICEBERG
COMMENT "Dimension de Clientes"
""")
```

### Dimension de Tiempo

```python
# - Crea tabla
spark.sql(f"""
CREATE TABLE IF NOT EXISTS {vSchemaGold}.DIM_DATES
(
  DATE_SK              BIGINT      NOT NULL,
  DATE_VALUE           TIMESTAMP   NOT NULL,
  DATE_NUM             INT,
  YEAR_MONTH_NUMBER    INT,
  CALENDAR_QUARTER     STRING,
  MONTH_NUM            INT,
  MONTH_NAME           STRING,
  MONTH_SHORT_NAME     STRING,
  WEEK_NUM             INT,
  DAY_NUM_OF_YEAR      INT,
  DAY_NUM_OF_MONTH     INT,
  DAY_NUM_OF_WEEK      INT,
  DAY_NAME             STRING,
  DAY_SHORT_NAME       STRING,
  QUARTER_NUM          INT,
  YEAR_QUARTER_NUM     INT,
  DAY_NUM_OF_QUARTER   INT
)
USING ICEBERG
PARTITIONED BY (YEAR_MONTH_NUMBER)
COMMENT "Dimension de Tiempo"
""")
```

### Dimension de Distribuidores

```python
# - Crea tabla
spark.sql(f"""
CREATE TABLE IF NOT EXISTS {vSchemaGold}.DIM_DEALERSHIPS
(
	DEALERSHIP_SK         BIGINT     NOT NULL COMMENT 'Llave Subrogada del distribuidor',
	DEALERSHIP_ID         INT        NOT NULL COMMENT 'Identificador del distribuidor',
	DEALERSHIP_MANAGER_ID INT                 COMMENT 'Identificador del gerente del distribuidor',
	DEALERSHIP_DESC       STRING              COMMENT 'Descripcion del distribuidor',
	DEALERSHIP_LOCATION   STRING              COMMENT 'Ciudad del distribuidor',
	DEALERSHIP_STATE      STRING              COMMENT 'Estado del distribuidor',
	DEALERSHIP_REGION     STRING              COMMENT 'Region del distribuidor',
	DEALERSHIP_COUNTRY    STRING              COMMENT 'Pais del distribuidor'
)
USING ICEBERG
COMMENT "Dimension de Distribuidores"
""")
```

### Dimension de Tipos de Pago

```python
# - Crea tabla
spark.sql(f"""
CREATE TABLE IF NOT EXISTS {vSchemaGold}.DIM_PAYMENTS
(
	PAYMENT_SK         BIGINT     NOT NULL COMMENT 'Llave Subrogada del tipo de pago',
	PAYMENT_ID         INT        NOT NULL COMMENT 'Identificador del tipo de pago',
	PAYMENT_TYPE_DESC  STRING              COMMENT 'Descripcion del tipo de pago'
)
USING ICEBERG
COMMENT "Dimension de Tipos de Pago"
""")
```

### Dimension de Productos

```python
# - Crea tabla
spark.sql(f"""
CREATE TABLE IF NOT EXISTS {vSchemaGold}.DIM_PRODUCTS
(
	PRODUCT_SK     BIGINT     NOT NULL COMMENT 'Llave Subrogada del producto',
	PRODUCT_ID     INT        NOT NULL COMMENT 'Identificador del producto',
	GROUP_ID       INT                 COMMENT 'Identificador del grupo al que pertenece el producto',
	PRODUCT_DESC   STRING              COMMENT 'Descripcion del producto',
	GROUP_DESC     STRING              COMMENT 'Descripcion del grupo al que pertenece el producto',
	DIVISION_DESC  STRING              COMMENT 'Descripcion de la division a la que pertenece el producto',
	SUPPLIER_DESC  STRING              COMMENT 'Descripcion del proveedor',
	COMPONENT_ID   STRING              COMMENT 'Identificador del componente',
	PRODUCT_COST   DECIMAL(12,2)       COMMENT 'Costo del producto'
)
USING ICEBERG
COMMENT "Dimension de Productos"
""")
```

### Dimension de Promociones

```python
# - Crea tabla
spark.sql(f"""
CREATE TABLE IF NOT EXISTS {vSchemaGold}.DIM_PROMOTIONS
(
	PROMO_SK    BIGINT     NOT NULL COMMENT 'Llave Subrogada de la promocion',
	PROMO_ID    INT        NOT NULL COMMENT 'Identificador de la promocion',
	PROMO_DESC  STRING              COMMENT 'Descripcion de la promocion',
	PROMO_TYPE  STRING              COMMENT 'Tipo de promocion',
	START_DK    INT                 COMMENT 'Fecha de inicio de la promocion, Formato YYYYMMDD',
	EXPIRY_DK   INT                 COMMENT 'Fecha de fin de la promocion, Formato YYYYMMDD',
	PROMO_COST  DECIMAL(12,2)       COMMENT 'Costo de la promocion',
	DISCOUNT    DECIMAL(12,2)       COMMENT 'Descuento de la promocion'
)
USING ICEBERG
COMMENT "Dimension de Promociones"
""")
```

### Hechos de Ventas

```python
# - Crea tabla
spark.sql(f"""
CREATE TABLE IF NOT EXISTS {vSchemaGold}.FACT_SALES
(
	DATE_SK        BIGINT         NOT NULL COMMENT 'Llave Subrogada de tiempo',
	CUST_SK        BIGINT         NOT NULL COMMENT 'Llave Subrogada del cliente',
	PRODUCT_SK     BIGINT         NOT NULL COMMENT 'Llave Subrogada del producto',
	DEALERSHIP_SK  BIGINT         NOT NULL COMMENT 'Llave Subrogada del distribuidor',
	PAYMENT_SK     BIGINT         NOT NULL COMMENT 'Llave Subrogada del tipo de pago',
	PROMO_SK       BIGINT         NOT NULL COMMENT 'Llave Subrogada de la promocion',
	UNITS_SOLD     INT                     COMMENT 'Unidades Vendidas',
	REVENUE        DECIMAL(15,2)           COMMENT 'Ganancia',
	COST           DECIMAL(15,2)           COMMENT 'Costo',
	DISCOUNT       DECIMAL(15,2)           COMMENT 'Descuento',
	HOLDBACK       DECIMAL(15,2)           COMMENT 'Retencion',
	REBATE         DECIMAL(15,2)           COMMENT 'Reembolso'
)
USING ICEBERG
PARTITIONED BY (DATE_SK)
COMMENT "Hechos de Ventas"
""")
```
